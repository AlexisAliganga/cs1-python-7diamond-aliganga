```python
# fa3prog3

name = input("Enter student's name: ")
score = float(input("Enter score: "))

if 90 <= score <= 100:
    grade = "A"
elif 80 <= score <= 89:
    grade = "B"
elif 70 <= score <= 79:
    grade = "C"
else:
    grade = "F"

print("\nStudent:", name)
print("Grade:", grade)
```
